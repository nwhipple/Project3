```python
import json
import pandas as pd
input_path = "./GlobalReactorList.json"
```

```python
json_data = {}
```

```python
with open(input_path,"r",encoding='utf-8') as file:
    json_data =json.load(file)
```

```python
json_data
```

```python
geojson = {
    "type":"FeatureCollection",
    "features":[]
}
```

```python
for feature in json_data["Data"]:
    geojson["features"].append({
      "type": "Feature",
      "geometry": {
        "type": "Point",
        "coordinates": [feature["Longitude"], feature["Latitude"]]
      },
      "properties": {**feature
      } 
    })
```

```python
geojson
```

```python
with open("reactorlist.geojson", "w", encoding="utf-8") as datafile:
    datafile.write(json.dumps(geojson))
```
